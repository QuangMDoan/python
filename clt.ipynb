# Central Limit Theorem
The theorem says that if you take large enough samples from a population and then
calculate the sample means, these means will be normally distributed.

- The theorem should hold as long as the sample size is large enough and the variable in question is independent and random

```python
# Generate random samples from a population
# take the means of the samples
# visualize the means
import helpers 
import numpy as np
%matplotlib inline

population_data = helpers.distribution(50, 10000)
helpers.histogram_visualization(population_data)
print(f'population mean {np.mean(population_data)}')
```

```python
def random_sample(population_data, sample_size):
    return np.random.choice(population_data, size = sample_size)

random_sample(population_data, 10)
```

```python
def sample_mean(sample):
    return np.mean(sample)

sample1 = random_sample(population_data, 10)

mean = sample_mean(sample1)
print(mean)
```

```python
# Use the random_sample() and sample_mean() functions to show how the central limit theorem works

# The code contains a for loop that draws a random sample of size N,
#   then takes the mean of the sample, and stores the mean in a list
#   each iteration of the loop will have different random sample
def central_limit_theorem(population, n, iterations):
    sample_mean_results = [sample_mean(random_sample(population, n)) for i in range(iterations)]
    return sample_mean_results

print("Means of all samples")
print(central_limit_theorem(population_data, 10, 1000)[:20])

```

```python
from matplotlib import pyplot as plt
def visualize_gaussian(sample_means, bins=100):
    plt.hist(sample_means, bins)
    plt.title('Histogram of the Sample Means')
    plt.xlabel('Mean Value')
    plt.ylabel('Count')

sample_means = central_limit_theorem(population_data, 10000, 10000)
visualize_gaussian(sample_means)
```
