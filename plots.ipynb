```python
def uniform_distribution_height(x_minimum, x_maximum):
    height = 1/(abs(x_maximum-x_minimum))
    return height


y = uniform_distribution_height(0, 360)
plt.plot([0,360], [y, y], color="blue")
plt.plot([0,0], [0,y], "--", color = "blue")
plt.plot([360,360], [0,y], "--", color = "blue")
plt.ylim([0, y+y/2])
plt.xlabel("Degrees")
plt.ylabel("Probability distribution")
plt.title("Uniform Probability Distribution \n for a Spinning Bottle")
plt.xticks(range(0, 361, 30))
plt.show()
```

```python
# Visualizing a Piece-Wise Continuous Probability Distribution
    # Write a function that, given a list of x-axis intervals, relative probabilities and a total probability
    # calculates the height of each bar. The sum of the area for all bars should be the total probability.

def bar_heights(intervals, relative_probabilities, total_probability):
    bar_areas = [x/sum(relative_probabilities) * total_probability for x in relative_probabilities]
    heights = [bar_areas[i]/(intervals[i+1]-intervals[i]) for i in range(len(intervals)-1)]
    return heights
```

```python
hour_intervals = [0, 5, 10, 16, 21, 24]
probability_intervals = [1, 5, 3, 6, 1/2]
accident_probability = 0.05

heights = bar_heights(hour_intervals, probability_intervals, accident_probability)

for i in range(len(probability_intervals)):
    x = [hour_intervals[i], hour_intervals[i+1]] 
    y = [heights[i], heights[i]]
    plt.plot(x, y, color='blue') 
    
    x = [x[0], x[0]]
    y = [0, heights[i]]
    plt.plot(x, y, '--', color='blue')

    x = [hour_intervals[i+1], hour_intervals[i+1]]
    plt.plot(x, y, '--', color='blue')

plt.xticks(range(0,25,1))
plt.xlabel('hours')
plt.ylabel('probability density function')
plt.title('Probability Density Function \n for Car Accidents During the Day')
plt.show()
```
